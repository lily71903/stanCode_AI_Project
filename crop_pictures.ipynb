```python
import os 
import pandas as pd
from PIL import Image
```

```python
df = pd.read_excel("fixed_data.xlsx", index_col=0)
df
```

```python
df['filename'] = [f.split(".")[0] for f in df['new_filename']]
df
```

### Avoid important information not in the photo

```python
for i in range(len(df)):
    
    for j in range(4):
        file_path = df.iloc[i]["path"]
        new_path = os.path.join("data", df.iloc[i]["filename"] + '-' + str(j) + '.jpg')
        print("{} => {}".format(file_path, new_path))

        try:
            img = Image.open(file_path)
            w, h = img.size
            h_w = [(0, 0, w, w), (0, (h-w)/4, w, (h-w)/4+w), (0, (h-w)/2, w, (h-w)/2+w), (0, (h-w)*3/4, w, 3*(h-w)/4+w)]
            w_h = [(0, 0, h, h), ((w-h)/4, 0, (w-h)/4+h, h), ((w-h)*3/4, 0, 3*(w-h)/4+h, h), ((w-h), 0, w, h)]
            if h > w:
                img_cropped = img.crop(h_w[j]) # L, T, R, B
            elif w > h:
                img_cropped = img.crop(w_h[j])
            else:
                pass
            img_resized = img_cropped.resize((1024, 1024))
            img_resized.save(new_path)

        except Exception as err:
            print("cannot handle {}".format(file_path))
            print(err)
```
